```python
df = pd.read_csv(data)
df=df.drop(['Unnamed: 0'],axis=1) #데이터 인덱스 번호 재설정
```

```python
df['대여일시'] = df['대여일시'].apply(lambda x: x.split(' ')[0]) #날짜와 시간 구분
df['반납일시'] = df['반납일시'].apply(lambda x: x.split(' ')[0]) #날짜와 시간 구분
df
```

```python
rental_place = pd.read_csv('/content/drive/MyDrive/ITSTUDY/Project1/csv/data/location.csv', encoding='cp949') #반납장소 데이터 파일 불러오기
rental_place2 = rental_place.copy()
rental_place2=rental_place2.drop([0,1,2,3]) #필요없는 행데이터 삭제
rental_place2=rental_place2.rename(columns={'대여소\n번호':'대여소 번호','Unnamed: 3':'주소', 'Unnamed: 4':'위도','Unnamed: 5':'경도'}) # 열 이름 재설정
rental_place2 = rental_place2.reset_index() #인덱스 정리
rental_place2=rental_place2.drop(['index','설치\n시기','설치형태','Unnamed: 8','운영\n방식'], axis=1) #인덱스 정리
rental_place2['대여소 번호'].astype=int #대여소 번호 숫자타입으로 변경
rental_place2.head()
```

```python
df.head()
```

```python
df_final = pd.merge(df,rental_place2,left_on='반납대여소번호', right_on='대여소 번호', how='left') # 전처리 완료한 두 데이터 결합 기준:반납 대여소 번호 기준
df_final=df_final.drop(['대여소 번호'],axis=1) #중복되어 표시되는 대여소 번호 삭제
df_final.rename(columns = {'대여 대여소명' : '대여장소', "반납대여소명":"반납장소","위도":"반납위도","경도":"반납경도",'소재지(위치)':'반납 자치구'}, inplace = True) #열 이름 재설정
#df_final[['대여장소','대여일시','이용거리', '이용시간','반납장소','반납대여소번호', '반납 자치구','주소','반납위도', '반납경도']]
df_final['반납위도']=df_final['반납위도'].astype('float') #위도, 경도 좌표 숫자타입으로 변경
df_final['반납경도']=df_final['반납경도'].astype('float')
final=df_final.copy()
```

```python
norental=final[final['반납위도'].isnull()] #지금은 폐쇄되거나 없어진 반납장소 탐색
```

```python
norental #약 529번의 반납이 기존에 반납장소에서 이루어짐
```

```python
norental['반납장소'].unique() #총 12군데의 반납장소가 사라짐
```

```python
final.dropna(inplace=True)  #반납장소 좌표를 찾을 수없는 곳들은 제외하고 우선 시각화 진행
```

```python
final=final.reset_index().drop(['index'],axis=1) #열 재설정
```

```python
gangseo = rental_place2[rental_place2['보관소(대여소)명'].isin(['강서구청'])]
gangnam = rental_place2[rental_place2['보관소(대여소)명'].isin(['강남구청'])]
region= pd.concat([gangnam, gangseo])
region=region.reset_index().drop(columns=['index']) #반납 좌표데이터에서 강남구청과 강서구청의 위도 경도 좌표 찾아오기
```
